### Make a DataFrame of Hadith narrators (Narrator, Hadith_Count). Find the top 5 narrators with the most hadiths.

```python
import requests
import pandas as pd
```

```python
narrators = requests.get("https://hadithapi.com/public/api/hadiths?apiKey=$2y$10$uUXLSww3EbKebq1cljtoOuXYgwhabHvoXBewf410XDyhDuqXQj1su").json()["hadiths"]["data"]
fetch = [f"{i['englishNarrator']}" for i in narrators]
df = pd.DataFrame(fetch, columns=["Narrator"])
df
```

```python
narrator_count = df.groupby("Narrator").size().reset_index(name="Hadith_Count")
top5 = narrator_count.sort_values(by="Hadith_Count", ascending=False).head(5)

print(top5)
```

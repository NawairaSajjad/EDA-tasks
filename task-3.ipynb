### Create a DataFrame with Surah_Number, Ayah_Count, Category (Makki/Madani). Show Surahs with more than 100 verses.


```python

import pandas as pd
import requests
```

```python
response = requests.get("https://api.alquran.cloud/v1/surah")
```

```python
if response.status_code == 200 :
    surah = response.json()["data"]
    df = pd.DataFrame(surah)
df
```

```python
df.columns
```

```python
df_specific = df[["number","numberOfAyahs","revelationType"]]
df_specific
```

```python
mask = df["numberOfAyahs"] > 100 
df[mask]
```
